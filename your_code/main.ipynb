#### Open the ` Employee.csv` file and name your dataset `employee`.

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1: Group By 

data_employee.groupby(['Department']).agg({'Salary':np.mean}).reset_index()

#If we want to calculate number of employyes data_employee.groupby(['Department']).agg({'Salary':len}).reset_index()
```

```python
# Method 2: Pivot Table

data_employee.pivot_table(index= ['Department'], values = ['Salary'])
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 1: Group By

data_employee.groupby(['Department','Title']).agg({'Salary':np.mean}).reset_index()
```

```python
# Method 2: Pivot 

data_employee.pivot_table(index= ['Department','Title'], values = ['Salary']).reset_index()
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
table = pd.pivot_table(data = data_employee, index=['Department','Title'], values=['Salary'], aggfunc=['mean','count']).reset_index()
table.head()

table.columns = ['Department', 'Title', 'MeanSalary', 'NumberEmployees']

table.head()
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
table1 = pd.pivot_table(data = data_employee, index=['Department','Title'], values=['Salary'], aggfunc=['count','max', 'min', 'mean', 'std']).reset_index()

table1.columns = ['Department', 'Title', 'NumberEmployees', 'MaxSalary', 'MinSalary', 'MeanSalary', 'StdSalary']
table1.head()

# We get NaN in rows where we have only 1 employee. This make sense because you cant calculate a measure of deviation where you only have 1 value and it is the mean
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
table2 = pd.pivot_table(data = data_employee, index=['Department'], values=['Salary'], aggfunc=['count','sum']).reset_index()
table2.columns = ['Department', 'NumberEmployees', 'TotalSalary']
table2.head(10)
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
table3 = pd.pivot_table(data = data_employee, index=['Department', 'Title'], values=['Salary', 'Years'], aggfunc={'Salary':'mean', 'Years':'sum'}).reset_index()

table3.head()
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
HR = data_employee[data_employee.Department == 'HR']# We create a DF with all the data from HR
IT = data_employee[data_employee.Department == 'IT']
Sales = data_employee[data_employee.Department == 'Sales']
HR = HR.drop(HR['Salary'].idxmax()) # Return index of the maximum element.
IT = IT.drop(IT['Salary'].idxmax())
Sales = Sales.drop(Sales['Salary'].idxmax())
data_employee_nomax = pd.concat([HR, IT, Sales]) # We concat all data without maximun

data_employee_nomax

# Now we create the pivot

table4 = pd.pivot_table(data=data_employee_nomax, index=['Department'], values=['Salary'], aggfunc=['mean']).reset_index()

table4.columns = ['Department', 'MeanSalary']

table4
```
